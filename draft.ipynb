```python
data['tourney_date'] = pd.to_datetime(data['tourney_date'], yearfirst=True, format='%Y%m%d')
```

```python
data_winner = data[['tourney_id', 'tourney_name', 'surface', 'tourney_level', 'tourney_date', 'match_num', 'winner_name', 'minutes', 'winner_rank']]
data_loser = data[['tourney_id', 'tourney_name', 'surface', 'tourney_level', 'tourney_date', 'match_num', 'loser_name', 'minutes', 'loser_rank']]
```

```python
data_winner.rename(columns={'winner_name':'player_name', 'winner_rank':'rank'}, inplace=True)
data_loser.rename(columns={'loser_name':'player_name', 'loser_rank':'rank'}, inplace=True)
```

```python
data_plays = pd.concat([data_winner, data_loser], axis=0)
```

```python
data_plays.reset_index(inplace=True, drop=True)
```
